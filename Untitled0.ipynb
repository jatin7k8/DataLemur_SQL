```python
def model_evaluation_metric(y_actual, y_pred, model):
  mse = round(mean_squared_error(y_true=y_actual, y_pred=y_pred),2)
  mae = round(mean_absolute_error(y_true=y_actual, y_pred=y_pred),2)
  rmse = round(mean_squared_error(y_true=y_actual, y_pred=y_pred, squared=False),2)
  print(f"MSE: {mse}\nMAE: {mae}\nRMSE: {rmse}")
```

```python
def acurracy_check(y_actual, y_pred, model):
  n = len(y_actual)
  k = len(model.coef_)
  r2 = round(r2_score(y_true=y_actual, y_pred=y_pred),2)
  adj_r2_score = round(1 - ((1-r2)*(n-1)/(n-k-1)),2)
  print(f"R2 Score: {r2}\nAdjusted R2 Score: {adj_r2_score}")
```

```python

print("Model Evaluation on Training Data:")
model_evaluation_metric(y_train, y_pred_train, model)
print()

print("Accuracy Check on Training Data:")
acurracy_check(y_train, y_pred_train, model)
print()


print("Model Evaluation on Test Data:")
model_evaluation_metric(y_test, y_pred_test, model)
print()

print("Accuracy Check on Test Data:")
acurracy_check(y_test, y_pred_test, model)
print()
```

# Homoscedasticity/No heterosadisity

The Breusch-Pagan test is a statistical test used to assess homoscedasticity in a regression model. It tests whether the variance of the residuals from a regression model is constant. Here's how you can perform the Breusch-Pagan test in Python:


* To test for homoscedasticity, there are several graphical and statistical methods that you can use:

  1. Residual plot
  2. Scatterplot

  3. Breusch-Pagan Test

  4. Goldfeld-Quandt Test

```python
# Check for heteroscedasticity
import statsmodels.stats.api as sms
# find error from ols
residuals = result.resid
stat, p_value,_,_ = sms.het_breuschpagan(residuals, result.model.exog)


print('Breusch-Pagan test statistic:', stat)
print('p-value:', p_value)

if p_value < 0.05:
    print("The test indicates heteroscedasticity.")
else:
    print("The test does not indicate heteroscedasticity.")
```

```python
# Plot residuals vs. fitted values for heteroscedasticity
plt.scatter(result.fittedvalues, residuals)
plt.xlabel('Fitted values')
plt.ylabel('Residuals')
plt.title('Residuals vs. Fitted Values')
plt.show()
```

# Normality check for residuals

## Summary of Conditions for Use
* Shapiro-Wilk Test: Use for small to moderate sample sizes, highly sensitive to normality deviations.
* Kolmogorov-Smirnov Test: Use for any sample size, flexible in comparing to any continuous distribution.
* Anderson-Darling Test: Use for all sample sizes, especially sensitive to deviations in the tails.
* Jarque-Bera Test: Use for large samples, focused on skewness and kurtosis.

```python
# Check for normality of residuals
stat, p_value ,_ , _= sms.jarque_bera(residuals)

print("Jarque-Bera Test Statistic:", stat)
print("p-value:", p_value)

if p_value < 0.05:
    print("Reject the null hypothesis: Residuals are not normally distributed.")
else:
    print("Fail to reject the null hypothesis: Residuals are normally distributed.")
```

# Correcting Heteroscedasticity
1. Transformations:
* Why: Transforming the dependent variable or some of the independent variables can stabilize the variance of residuals.
* When to Use: Use transformations when you observe a clear pattern of heteroscedasticity in the residual plot or when the Breusch-Pagan test suggests significant heteroscedasticity.

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Add a constant to the model (intercept term)
X_train_const = sm.add_constant(X_train)
X_test_const = sm.add_constant(X_test)

# Fit the OLS model
model = sm.OLS(y_train, X_train_const).fit()

# Predict and calculate residuals
y_train_pred = model.predict(X_train_const)
residuals = y_train - y_train_pred

# Check for heteroscedasticity using Breusch-Pagan test
stat, p_value,_,_ = sms.het_breuschpagan(residuals, X_train_const)

print('Breusch-Pagan test statistic:', stat)
print('p-value:', p_value)

if p_value < 0.05:
    print("The test indicates heteroscedasticity.")
else:
    print("The test does not indicate heteroscedasticity.")
```
